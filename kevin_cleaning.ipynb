```python
import pandas as pd
import csv
```

```python
csv = pd.read_csv('../top10cities_ca_tx.csv')
```

```python
csv['Start_Time'] = pd.to_datetime(csv['Start_Time'])
csv['End_Time'] = pd.to_datetime(csv['End_Time'])
csv['Zipcode'] = csv['Zipcode'].str[:5]
```
